We first need to create a folium `Map` object, with an initial center location to be NASA Johnson Space Center at Houston, Texas.

We could use `folium.Circle` to add a highlighted circle area with a text label on a specific coordinate. For example,

```python
# Initial the map
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# For each launch site, add a Circle object based on its coordinate (Lat, Long) values. In addition, add Launch site name as a popup label

# Launch site coordinates in a DataFrame
launch_sites = pd.DataFrame({
    'Launch Site': ['CCAFS LC-40', 'CCAFS SLC-40', 'KSC LC-39A', 'VAFB SLC-4E'],
    'Lat': [28.562302, 28.563197, 28.573255, 34.632834],
    'Long': [-80.577356, -80.576820, -80.646895, -120.610745]
})

# Add circles and markers for each launch site
for index, site in launch_sites.iterrows():
    coordinate = [site['Lat'], site['Long']]
    launch_site = site['Launch Site']
    
    # Add a circle for each launch site
    folium.Circle(
        location=coordinate,
        radius=1000,
        color='#000000',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.4
    ).add_child(folium.Popup(launch_site)).add_to(site_map)
    
    # Add a text marker for each launch site
    folium.map.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % launch_site
        )
    ).add_to(site_map)

# Display the map
site_map
```

```python
# Task 2: Mark the success/failed launches for each site on the map
# Initialize the map centered at NASA JSC
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# Define success and failure colors
success_color = 'green'
failure_color = 'red'

# Add markers for launch outcomes
for index, row in spacex_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    outcome = 'Success' if row['class'] == 1 else 'Failure'
    color = success_color if row['class'] == 1 else failure_color

    # Add a marker for each launch with a popup label and colored icon
    folium.Marker(
        location=coordinate,
        icon=folium.Icon(color=color, icon='ok-circle' if row['class'] == 1 else 'remove-circle'),
        popup=folium.Popup(f"Launch Site: {row['Launch Site']}<br>Outcome: {outcome}", max_width=300)
    ).add_to(site_map)

# Display the map
site_map

```

Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red

# Define a function to determine marker color based on class value
def assign_marker_color(class_value):
    if class_value == 1:
        return 'green'  # Success
    else:
        return 'red'  # Failure

# Apply the function to create a new column 'marker_color'
spacex_df['marker_color'] = spacex_df['class'].apply(assign_marker_color)

# Display the updated DataFrame
spacex_df.head()
```

*TODO:* For each launch result in `spacex_df` data frame, add a `folium.Marker` to `marker_cluster`


```python
# Add marker_cluster to current site_map
site_map.add_child(marker_cluster)

# Iterate through the DataFrame to add markers
for index, record in spacex_df.iterrows():
    # Extract coordinates for the marker
    coordinate = [record['Lat'], record['Long']]
    
    # Create a Marker with icon color based on 'marker_color' column
    marker = folium.Marker(
        location=coordinate,
        icon=folium.Icon(color=record['marker_color']),  # Use 'marker_color' for color
        popup=f"Launch Site: {record['Launch Site']}, Outcome: {'Success' if record['class'] == 1 else 'Failure'}"
    )
    
    # Add the marker to the marker cluster
    marker_cluster.add_child(marker)

# Display the map
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline

# Coordinates for the CCAFS LC-40 launch site
launch_site_lat, launch_site_lon = 28.562302, -80.577356

# Coordinates for the closest coastline point
coastline_lat, coastline_lon = 28.56169, -80.56768

# Calculate the distance between the launch site and the coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
print("Distance to closest coastline: {:.2f} km".format(distance_coastline))


```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map

# Add a marker for the closest coastline with the calculated distance
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12px; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline),
    )
)

# Add the marker to the site map
site_map.add_child(distance_marker)

# Display the updated map
site_map
```

*TODO:* Draw a `PolyLine` between a launch site to the selected coastline point


```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate

launch_site_coords = [28.562302, -80.577356]
coastline_coords = [28.56169, -80.56768]

# Create a PolyLine using the coordinates
lines = folium.PolyLine(locations=[launch_site_coords, coastline_coords], weight=2, color='blue')

# Add the PolyLine to the map
site_map.add_child(lines)

# Display the updated map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
```

```python
# Define coordinates for the CCAFS LC-40 launch site and proximity points
launch_site_coords = [28.562302, -80.577356]
coastline_coords = [28.56169, -80.56768]
railroad_coords = [28.56212, -80.58716]
highway_coords = [28.56246, -80.57064]
city_center_coords = [28.53599, -81.38672]

# Calculate distances to each proximity point
distance_coastline = calculate_distance(*launch_site_coords, *coastline_coords)
distance_railroad = calculate_distance(*launch_site_coords, *railroad_coords)
distance_highway = calculate_distance(*launch_site_coords, *highway_coords)
distance_city = calculate_distance(*launch_site_coords, *city_center_coords)

# Initialize the map
site_map = folium.Map(location=launch_site_coords, zoom_start=10)

# Draw PolyLines for each proximity point
folium.PolyLine([launch_site_coords, coastline_coords], weight=2, color='blue').add_to(site_map)
folium.PolyLine([launch_site_coords, railroad_coords], weight=2, color='red').add_to(site_map)
folium.PolyLine([launch_site_coords, highway_coords], weight=2, color='green').add_to(site_map)
folium.PolyLine([launch_site_coords, city_center_coords], weight=2, color='purple').add_to(site_map)

# Add Markers with distances for each proximity point
folium.Marker(
    coastline_coords,
    popup=f"Closest Coastline: {distance_coastline:.2f} km",
    icon=DivIcon(icon_size=(20, 20), icon_anchor=(0, 0), 
                 html=f'<div style="font-size: 12; color:blue;"><b>{distance_coastline:.2f} KM</b></div>')
).add_to(site_map)

folium.Marker(
    railroad_coords,
    popup=f"Closest Railroad: {distance_railroad:.2f} km",
    icon=DivIcon(icon_size=(20, 20), icon_anchor=(0, 0), 
                 html=f'<div style="font-size: 12; color:red;"><b>{distance_railroad:.2f} KM</b></div>')
).add_to(site_map)

folium.Marker(
    highway_coords,
    popup=f"Closest Highway: {distance_highway:.2f} km",
    icon=DivIcon(icon_size=(20, 20), icon_anchor=(0, 0), 
                 html=f'<div style="font-size: 12; color:green;"><b>{distance_highway:.2f} KM</b></div>')
).add_to(site_map)

folium.Marker(
    city_center_coords,
    popup=f"City Center (Orlando): {distance_city:.2f} km",
    icon=DivIcon(icon_size=(20, 20), icon_anchor=(0, 0), 
                 html=f'<div style="font-size: 12; color:purple;"><b>{distance_city:.2f} KM</b></div>')
).add_to(site_map)

# Add a marker for the launch site
folium.Marker(
    launch_site_coords,
    popup="CCAFS LC-40 Launch Site",
    icon=folium.Icon(color='orange', icon='star')
).add_to(site_map)

# Display the updated map
site_map
```
